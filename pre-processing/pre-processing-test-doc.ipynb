```python
def is_valid_time(s):
    h = int(s[:2])
    m = int(s[3:])
    if h > 9 and h < 16:
        return True
    elif h == 9 and m >= 30:
        return True
    elif h == 16 and m == 0:
        return True
    else:
        return False
```
